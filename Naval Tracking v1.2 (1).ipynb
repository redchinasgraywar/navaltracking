```python
for i in range(len(keywords)):
    kword = keywords[i]
    print('\033[91m' + '\033[1m' + kword + '\033[0;0m')

    for n in range(5):
      r = requests.get(search_links[i*3 + n])
      soup = bs(r.content, 'html.parser')

      links = []
      links_all = soup.find_all('a', class_= 'uk-button box-button', href=True)
      for a in links_all:
          links.append('https://www.dvidshub.net' + a['href'])

      for a in links:
          r1 = requests.get(a)
          soup1 = bs(r1.content, 'html.parser')
          loc = soup1.find_all('td')
          title = soup1.find_all('title')
          try:
              output = str(title[0])[7:-8] + '\n' + str(loc[13])[4:-5] + ' ' + str(loc[15])[4:-5] + ' ' + str(loc[11])[4:-5] + ' ' + str(loc[1])[4:-5] + '\n' + a + '\n'
          except:
              try:
                  output = str(title[0])[7:-8] + '\n' + str(loc[7])[4:-5] + ' ' + str(loc[1])[4:-5] + '\n' + a + '\n'
              except:
                  output = 'Unavailable'
          print(output)
    print('\n \n \n \n')
```
